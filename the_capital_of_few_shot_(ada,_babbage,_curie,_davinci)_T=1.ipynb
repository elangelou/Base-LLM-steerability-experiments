```python
!pip install openai
!pip install backoff

import openai
import backoff

OPENAI_API_KEY = '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'
openai.api_key = OPENAI_API_KEY

def generate_question(i):
    templates =[
        "The capital of France is Paris, The capital of Italy is Rome, the capital of Spain is ",
        "The capital of Germany is Berlin, The capital of Russia is Moscow, the capital of Spain is ",
        "The capital of Brazil is Brasília, The capital of Argentina is Buenos Aires, the capital of Spain is ",
        "The capital of Japan is Tokyo, The capital of India is New Delhi, the capital of Spain is ",
        "The capital of Australia is Canberra, The capital of New Zealand is Wellington, the capital of Spain is ",
        "The capital of Canada is Ottawa, The capital of Mexico is Mexico City, the capital of Spain is ",
        "The capital of Egypt is Cairo, The capital of South Africa is Pretoria, the capital of Spain is ",
        "The capital of China is Beijing, The capital of South Korea is Seoul, the capital of Spain is ",
        "The capital of United Kingdom is London, The capital of Portugal is Lisbon, the capital of Spain is ",
        "The capital of Turkey is Ankara, The capital of Bulgaria is Sofia, the capital of Spain is ",

    ]
    question = templates[i]
    answer = "Madrid"
    return question, answer


@backoff.on_exception(backoff.expo,Exception)
def ask_models(question):
    model_names = ["ada", "babbage", "curie", "davinci"]
    responses = {}
    for model in model_names:
        response = openai.Completion.create(
            engine=model,
            prompt=question,
            temperature=1,
            max_tokens=5
        )
        responses[model] = response.choices[0].text.strip()
    return responses

def main():
    correct_answers = {'ada': 0, 'babbage': 0, 'curie': 0, 'davinci': 0}
    total_questions = 10

    for i in range(total_questions):
        question, answer = generate_question(i)
        print(f"Question {i + 1}: {question}")
        model_responses = ask_models(question)

        for model, response in model_responses.items():
            print(f"{model} answered: {response}")
            if response == answer:
                correct_answers[model] += 1
        print()

    for model, count in correct_answers.items():
        print(f"{model}: {count} correct answers out of {total_questions}")

if __name__ == "__main__":
     main()
```
