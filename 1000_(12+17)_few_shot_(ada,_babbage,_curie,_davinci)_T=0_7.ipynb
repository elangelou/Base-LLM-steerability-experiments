```python
!pip install openai
!pip install backoff

import openai
import backoff


openai_api_key = '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'
openai.api_key = openai_api_key

def generate_question(i):
    templates =[
        "12+1=13, 12+5=17, 12+7=19, 12+17=",
        "4+8=12, 5+7=12, 7+5=12, 12+17= ",
        "12+3=15, 12+6=18, 12+9=21, 12+17= ",
        "6+6=12, 8+4=12, 9+3=12, 12+17= ",
        "12+1=13, 12+4=16, 12+8=20, 12+17= ",
        "12+5=17, 12+6=18, 12+11=23, 12+17= ",
        "12+0=12, 12+12=24, 12+15=27, 12+17= ",
        "1+11=12, 4+8=12, 6+6=12, 12+17= ",
        "11+1=12, 10+2=12, 12+5=17, 12+17= "

    ]
    question = templates[i % len(templates)]
    answer = 12 + 17
    return question, answer

@backoff.on_exception(backoff.expo,Exception)
def ask_models(question):
    model_names = ["ada", "babbage", "curie", "davinci"]
    responses = {}
    for model in model_names:
        response = openai.Completion.create(
            engine=model,
            prompt=question,
            temperature=0.7,
            max_tokens=10
        )
        responses[model] = response.choices[0].text.strip()
    return responses

def main():
      correct_answers = {'ada': 0, 'babbage': 0, 'curie': 0, 'davinci': 0}
      total_questions = 1000
      questions_string = ""
      model_responses_all = {'ada': [], 'babbage': [], 'curie': [], 'davinci': []}

      for i in range(total_questions):
          question, answer = generate_question(i)
          print(f"Question {i +1}: {question}")
          questions_string += question + "\n"
          model_responses = ask_models(question)

          for model, response in model_responses.items():
              model_responses_all[model].append(response)
              try:
                  if int(response) == answer:
                      correct_answers[model] += 1
              except ValueError:
                  pass

      print(questions_string)
      for model, responses in model_responses_all.items():
          print(f"{model} responses:")
          print("\n".join(responses))
          print()

      for model, count in correct_answers.items():
        print(f"{model}: {count} correct answers out of {total_questions}")

if __name__ == "__main__":
    main()
```
